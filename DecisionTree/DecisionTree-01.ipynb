**Decision tree** builds regression or classification models in the form of a tree structure. It breaks down a dataset into smaller and smaller subsets while at the same time an associated decision tree is incrementally developed.A decision tree is arriving at an estimate by asking a series of questions to the data, each question narrowing our possible values until the model get confident enough to make a single prediction.

## 1. Decision Tree Classifier

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
iris = load_iris()
X = iris.data[:, 2:] # petal length and width 
y = iris.target
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(tree_clf, 
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

* **Gini index** varies between values 0 and 1, where 0 expresses the purity of classification, i.e. All the elements belong to a specified class or only one class exists there. And 1 indicates the random distribution of elements across various classes. The value of 0.5 of the Gini Index shows an equal distribution of elements over some classes.


Where Pi denotes the probability of an element being classified for a distinct class.

```python
tree_clf.predict([[5, 1.5]]) ## Predict versicolor
```

## 2. Descision Tree Regressor

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
```

```python
# Prepare the data data
boston = datasets.load_boston()
X = boston.data
y = boston.target
```

```python
X[0]
```

```python
regr = DecisionTreeRegressor(max_depth=3, random_state=1234)
model = regr.fit(X, y)
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(regr, feature_names=boston.feature_names, filled=True)
```

* Decision trees regression normally use mean squared error (MSE) to decide to split a node in two or more sub-nodes. 
* For each subset, it will calculate the MSE separately. The tree chooses the value with results in smallest MSE value.
